```python
import os

# Set the directory path where the images are located
directory = r"D:\archive\train\athletic_field-outdoor"
# Loop through the range of image names you want to delete
for i in range(101, 5001):
    filename = f"{i:08d}.jpg"
    filepath = os.path.join(directory, filename)
    
    # Check if the file exists
    if os.path.exists(filepath):
        # Delete the file
        os.remove(filepath)
        print(f"Deleted: {filename}")
    else:
        print(f"File not found: {filename}")
```
