```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
query = "narendramodi until:2023-02-10 since:2023-01-01"
tweets = []
limit = 2000
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content,tweet.likeCount])
        tweets.append([tweet.date, tweet.username, tweet.content,tweet.likeCount])

df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet','Like'])
print(df)
# to save to csv
# df.to_csv('tweets.csv')
```
